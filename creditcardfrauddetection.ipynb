# Credit Card Fraud Detection 

### Problem statement

According to 2016 data released by ACI Worldwide and financial industry consultant Aite Group, nearly 1 in 3 consumers globally have been victimized by card fraud in the past five years. Card fraud losses are a huge problem. The Nilson Report, a publication covering global payment systems, reported recently that global card fraud losses equaled $22.8 billion in 2016, an increase of 4.4 percent over 2015. 

Leveraging Machine learning and Deep learning techniques will enable payment vendors to identify fradulent transaction and can come up with counter measures to stop them and can save a lot of money.

This is a classification problem since the transaction is going to be fradulent or not. The dataset I used for the below problem is a real dataset but anonymized for privacy reasons and the same can be downloaded on Kaggle. Please follow the below link to download the dataset.

https://www.kaggle.com/mlg-ulb/creditcardfraud/downloads/creditcardfraud.zip/3


We shall start solving the problem by importing python libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import keras

np.random.seed(2)
```

```python
data = pd.read_csv('creditcard.csv')
```

Data Exploration

```python
data.head(10)
```

### Data pre-processing

```python
from sklearn.preprocessing import StandardScaler
data['normalizedAmount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
data = data.drop(['Amount'], axis = 1)
```

```python
data.head()
```

```python
data = data.drop(['Time'],axis = 1)
data.head()
```

```python
X = data.iloc[:, data.columns != 'Class']
y = data.iloc[:, data.columns == 'Class']
```

```python
X.head()
```

```python
y.head()
```

Splitting the dataset into training and test dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

### Deep Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

Creating a dense deep neural network which is sequential in model. We are going to use dropout layers in between so as to prevent overfitting. We will be using Relu activation function for all dense layers except the last one where we will be using sigmoid activation function as we are going to determine the transaction is fradulent or not which is a binary classification problem here.

We will reiterate the algorithm once it is modelled and will evaluate if we require more dense model

```python
model = Sequential([
    Dense(units = 16,input_dim = 29,activation = 'relu'),
    Dense(units = 24,activation = 'relu'),
    Dropout(0.5),
    Dense(units = 20,activation = 'relu'),
    Dense(units = 24,activation = 'relu'),
    Dense(units = 1,activation = 'sigmoid'),
])
```

```python
model.summary()
```

Training the model

```python
model.compile(optimizer='adam',loss ='binary_crossentropy',metrics = ['accuracy'])
model.fit(X_train,y_train,batch_size=15,epochs=5)
```

```python
score = model.evaluate(X_test, y_test)
```

```python
print(score)
```

```python
y_pred = model.predict(X_test)
y_test = pd.DataFrame(y_test)
```
